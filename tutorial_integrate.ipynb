### Step 5: Models splitting
Each cell relative model needs to be converted to a mat file in order to exploit the MATLAB function to convert model from the reversible into the irreversible format.

````python
lcellLines = ['MCF102A', 'SKBR3', 'MCF7', 'MDAMB231', 'MDAMB361']
for cellLine in lcellLines:
    model = cb.io.read_sbml_model(os.path.join(MODELDIR, 'ENGRO2_' + cellLine + '.xml'))
    cb.io.save_matlab_model(model, os.path.join(MODELDIR, 'ENGRO2_' + cellLine + '.mat'))
````

The following MATLAB functions are then exploited to split models into their irreversible format:
```matlab
model = readCbModel('./models/ENGRO2_MCF102A.mat')
[modelIrrev, matchRev, rev2irrev, irrev2rev] = convertToIrreversible(model)
outmodel = writeCbModel(modelIrrev, 'format','sbml', 'fileName', './models/ENGRO2_MCF102A_wIrrRxns.xml')
```
